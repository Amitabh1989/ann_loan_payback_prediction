```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
non_top_1_perc.columns
```

```python
X = non_top_1_perc.drop('price', axis=1).values
```

```python
y = non_top_1_perc['price'].values
```

```python
scalar = MinMaxScaler()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
scalar.fit(X_train)
```

```python
X_train = scalar.transform(X_train)
X_train
```

```python
X_test = scalar.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Activation
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
'''Choosing an optimizer and lossÂ¶
Keep in mind what kind of problem you are trying to solve:

# For a multi-class classification problem
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# For a binary classification problem
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# For a mean squared error regression problem
model.compile(optimizer='rmsprop',
              loss='mse')'''
```

```python
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, verbose=1, validation_data=(X_test,y_test),
          batch_size=128, epochs=400)
```

```python
loss = model.history.history
```

```python
loss
```

```python
loss = pd.DataFrame(loss)
```

```python
loss
```

```python
plt.figure(figsize=(14,14))
loss.plot(figsize=(14,14))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
print("MSE   : %s" % mean_squared_error(y_test, y_pred))
print("RSE   : %s" % (mean_squared_error(y_test, y_pred)**0.5))
print("MAE   : %s" % mean_absolute_error(y_test, y_pred))
print("EVS   : %s" % explained_variance_score(y_test, y_pred))
```
