```python
from sklearn.model_selection import train_test_split

X = df.drop('benign_0__mal_1', axis=1)
y = df['benign_0__mal_1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test = scalar.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(10,10))
loss.plot(figsize=(10,10))
```

```python
from tensorflow.keras.layers import Dropout
```

```python
# Redefine model

model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(10,12))
loss.plot()
```

```python
# Redefine model

model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')

model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), epochs=600, callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot()
```

```python
pred = model.predict(X_test)
pred
9.9045837e-01
```

```python
pred = (model.predict(X_test)>0.5).astype('int32')
pred
```

```python
predictions =(model.predict(X_test)>0.5).astype('int32')
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, classification_report, confusion_matrix

print("MSE   : %s" % (mean_squared_error(y_test, predictions)))
print("MAE   : %s" % (mean_absolute_error(y_test, predictions)))
print("CLR   : %s" % (classification_report(y_test, predictions)))
```

```python
print("CFM   : %s" % (confusion_matrix(y_test, predictions)))
```
